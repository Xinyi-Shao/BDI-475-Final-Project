### 5. Find the total number of confirmed cases at 5/6/21

### 8. Filter rows with >=50,000 recovered cases at 5/6/21

To filter the rows with >=50,000 recovered cases at 5/6/21, I will create a new variable, df_r5, and store the matched result in it using the data from df_r.

### 9. Filter France province with confirmed COVID-19 cases lower than 5,000 at 5/6/21

To filter France provinces with confirmed COVID-19 cases lower than 5,000 at 5/6/21, I will create a new variable, df_fr_5k, and store the matched result in it using the df dataset.
After filtering the rows, I will then create a boxplot to show the data's distribution.

```python
df_fr_5k = df[(df['Country/Region'] == 'France') & (df['5/6/21'] <= 5000)]

df_fr_5k.head(10)
```

```python
fig = px.box(df[(df['Country/Region'] == 'France') & (df['5/6/21'] <= 5000)],
             x='5/6/21',
             orientation='h',
             title='France province with confirmed COVID-19 cases lower than 5,000')
fig.show()
```

### 12. Create a summary of metrics of confirmed cases

To create a summary of metrics of confirmed cases, I will create a new variable named df_metrics.

```python
df_metrics = df.groupby('5/6/21', as_index=False)

df_metrics.head(10)
```

### 13. Sort by number of confirmed cases at 5/6/21

To sort by number of confirmed cases at 5/6/21 from the highest to lowest, I will create a new variable, df_by_cc, sort the values in descending order, and store the result in it.

After sorting the data, I will use a bar chart to show the 20 country/region with the highest number of confirmed cases at that date.



```python
df_by_cc = df.sort_values('5/6/21', ascending=False)

df_by_cc.head(10)
```

```python
fig = px.bar(
    df_by_cc.head(20),
    x='5/6/21',
    y='Country/Region',
    color='Country/Region',
    template='plotly_dark',
    title='Confirmed Cases Breakdown by Country/Region',
    height=600
)

fig.update_yaxes(categoryorder='total ascending')

fig.show()
```

### 14. Sort by number of death resulted by COVID-19 at 5/6/21

To sort number of death resulted by COVID-19 at 5/6/21, I will create a new variable, df_by_nd, sort the values in descending order, and store the result in it.

After sorting the data, I will use a bar chart to show the 20 country/region with the highest number of death number resulted by COVID-19 at that date.

```python
df_by_nd = df_d.sort_values('5/6/21', ascending=False)

df_by_nd.head(10)
```

```python
fig = px.bar(
    df_by_nd.head(20),
    x='5/6/21',
    y='Country/Region',
    color='Country/Region',
    template='plotly_dark',
    title='COVID-19 Death Number Breakdown by Country/Region',
    height=600
)

fig.update_yaxes(categoryorder='total ascending')

fig.show()
```

### 15. Sort by recovered COVID-19 cases at 5/6/21

To sort recovered COVID-19 cases at 5/6/21, I will create a new variable named df_by_rc, sort the values in descending order, and store the result in it.

After sorting the data, I will use a bar chart to show the 30 country/region with the highest number of recovered COVID-19 cases at that date.

```python
df_by_rc = df_r.sort_values('5/6/21', ascending=False)

df_by_rc.head(10)
```

```python
fig = px.treemap(
    df_by_rc.head(30),
    path=['Country/Region'],
    title='COVID-19 Recovered Number Breakdown by Country/Region',
    values='5/6/21',
    height=700
)

fig.show()
```
