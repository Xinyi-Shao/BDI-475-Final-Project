**BDI 475 Final Project - COVID-19 Data Analysis**

I chose to analyze these data sets of COVID-19 because I want to gain a deeper insight into the current situation of global pendamic by filtering and sorting the confirmed number of cases, the death number resulted by COVID-19, and the number of recovered cases of each counry/region.

1. Import Python packages

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
pd.set_option('display.max_columns', 50)
```

2. import data sets

```python
df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
```

```python
df_d = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
```

```python
df_r = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
```

3. Fine the number of rows and and columns in each data set

```python
num_rows_c = df.shape[0]
num_cols_c = df.shape[1]
print(f'There are {num_rows_c} rows and {num_cols_c} columns in the COVID-19 global confirmed cases dataset.')
```

```python
num_rows_d = df_d.shape[0]
num_cols_d = df_d.shape[1]
print(f'There are {num_rows_d} rows and {num_cols_d} columns in the COVID-19 global death number dataset.')
```

```python
num_rows_r = df_r.shape[0]
num_cols_r = df_r.shape[1]
print(f'There are {num_rows_r} rows and {num_cols_r} columns in the COVID-19 global recovery cases dataset.')
```

5. Find the total number of confirmed cases at 5/6/21

```python
total_num_confirmed = df['5/6/21'].sum()

print(f'{total_num_confirmed} COVID-19 cases were confirmed at 5/6/21.')
```

```python
total_num_deaths = df_d['5/6/21'].sum()
print(f'{total_num_deaths} cases of death resulted by COVID-19 were confirmed at 5/6/21.')
```

```python
total_num_recover = df_r['5/6/21'].sum()
print(f'{total_num_recover} recovered COVID-19 cases were confirmed at 5/6/21.')
```

6. Filter rows with 0 confirmed cases

```python
df_0 = df[df['5/6/21'] == 0]

display(df_0.head(10))
print(f'There are {df_0.shape[0]} countries/regions with 0 confirmed cases at 5/6/21.')
```

7. Find rows with 0 number of death resulted by COVID-19

```python
df_d0 = df_d[df_d['5/6/21'] == 0]

display(df_d0.head(10))
print(f'There are {df_d0.shape[0]} countries/regions with 0 number of death resulted by COVID-19 at 5/6/21.')
```

8. Filter rows with >=50,000 recovered cases

```python
df_r5 = df_r[df_r['5/6/21'] >=50000]

display(df_r5.head(10))
print(f'There are {df_r5.shape[0]} countries/regions with greater than or equal to 50,000 recovered COVID-19 cases at 5/6/21.')
```

9. Filter France province with confirmed COVID-19 cases lower than 5,000

```python
df_fr_5k = df[(df['Country/Region'] == 'France') & (df['5/6/21'] <= 5000)]

df_fr_5k.head(10)
```

10. Filter China province with confirmed number of death resulted by COVID-19 cases lower than or equal to 500

```python
df_ch_5 = df_d[(df_d['Country/Region'] == 'China') & (df_d['5/6/21'] <= 500)]

df_ch_5.head(10)
```

11. Filter India province with confirmed recovered COVID-19 cases greater than or equal to 500

```python
df_in_5k = df_r[(df_r['Country/Region'] == 'India') & (df_r['5/6/21'] >= 500)]

df_in_5k.head(10)
```

11. Create a summary of metrics of confirmed cases

```python
df_metrics = df.groupby('5/6/21', as_index=False)

df_metrics.head(10)
```

12. Sort by number of confirmed cases at 5/6/21

```python
df_by_cc = df.sort_values('5/6/21', ascending=False)

df_by_cc.head(10)
```

```python
fig = px.bar(
    df_by_cc,
    x='5/6/21',
    y='Country/Region',
    color='Country/Region',
    template='plotly_dark',
    title='Confirmed Cases Breakdown by Country/Region',
    height=600
)

fig.update_yaxes(categoryorder='total ascending')

fig.show()
```

13. Sort by number of death resulted by COVID-19 at 5/6/21

```python
df_by_nd = df_d.sort_values('5/6/21', ascending=False)

df_by_nd.head(10)
```

```python
fig = px.bar(
    df_by_nd,
    x='5/6/21',
    y='Country/Region',
    color='Country/Region',
    template='plotly_dark',
    title='COVID-19 Death Number Breakdown by Country/Region',
    height=600
)

fig.update_yaxes(categoryorder='total ascending')

fig.show()
```

14. Sort by recovered COVID-19 cases at 5/6/21

```python
df_by_rc = df_r.sort_values('5/6/21', ascending=False)

df_by_rc.head(10)
```

```python
fig = px.pie(
    df_by_rc,
    names='Country/Region',
    title='Recovery Cases Country/Region breakdown',
    width=800,
    height=700
)

fig.show()
```
